```python
open_excel_ensani=pd.read_excel('ریاضی-v1.xlsx', names=['p','name','id','purl','vurl'])
open_excel_ensani
```

```python
import requests 
import time
for i in open_excel_ensani.itertuples():
    dic={'Authorization':'apikey 9c7ea2c7-9587-5e17-95a7-b370363a8aff'}
    if isinstance(i.id,str) and ('-' in i.id):
        print(i.id)
        time.sleep(0.3)
        r = requests.get('https://napi.arvancloud.ir/vod/2.0/videos/'+i.id,headers=dic)
        if 'data' in r.json():
            open_excel_ensani.at[i.Index,'purl']=r.json()['data']['player_url']
            open_excel_ensani.at[i.Index,'vurl']=r.json()['data']['video_url']
    
```
